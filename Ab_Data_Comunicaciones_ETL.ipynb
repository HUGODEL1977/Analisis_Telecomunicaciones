ETL Acceso a Internet

```python
import pandas as pd
import numpy as np

# Define la ruta completa del archivo CSV
Acceso_ba_ba_prov = 'Data_Acceso_Internet\Data_Acceso_Internet_Archivos_ Iniciales\Accesos a banda ancha y banda angosta por provincia.csv'  # Reemplaza 'ruta/a/tu/archivo.csv' con la ruta real

# Convertimos nuestro archivo en un df
df1I = pd.read_csv(Acceso_ba_ba_prov)

# Detectar valores NaN o nulos en el DataFrame y contar la cantidad de valores NaN o nulos en cada columna
df5I= df1I.isna().sum() 

# Verifica si hay valores NaN en la columna "Dial up"
if df1I['Dial up'].isna().any():
    # Si hay valores NaN, puedes llenarlos con ceros o cualquier otro valor predeterminado si lo deseas.
    # En este caso, llenaremos los valores NaN con ceros.
    df1I['Dial up'].fillna(0, inplace=True)

# Convierte la columna "Dial up" a tipo de dato entero
df1I['Dial up'] = df1I['Dial up'].astype(int)
# Convierte la columna 'Año' en una columna de fecha
df1I['Año'] = pd.to_datetime(df1I['Año'], format='%Y')

# Sumar las columnas "Banda ancha fija" y "Dial up" y guardar el resultado en la columna "Total"
df1I['Total'] = df1I['Banda ancha fija'] + df1I['Dial up']

# Muestra el DataFrame con la columna "Total" actualizada
pd.set_option('display.expand_frame_repr', False)
print(df1I.head(10))

data_types = df1I.dtypes
print(data_types)

# Define la ruta donde quieres guardar el archivo CSV
ruta_csv = 'Accesos a banda ancha y banda angosta por provincia_cleaned.csv'  # Reemplaza con la ruta y el nombre que desees

# Guarda el DataFrame en un archivo CSV
df1I.to_csv(ruta_csv, index=False)
```

```python
# Define la ruta completa del archivo CSV
Listado_loc_ci = 'Data_Acceso_Internet\Data_Acceso_Internet_Archivos_ Iniciales\Listado de localidades con conectividad a internet.csv'  # Reemplaza 'ruta/a/tu/archivo.csv' con la ruta real

df3I = pd.read_csv(Listado_loc_ci)

df3I = df3I.replace('--', np.nan)
df3I = df3I.drop('Partido', axis=1)
df3I = df3I.drop_duplicates()
pd.set_option('display.expand_frame_repr', False)

df3I = df3I.fillna(False) # Se reemplazan los espacios vacios por false que indica que esta localidad
# no tiene este tipo de conectividad a Internet

# Lista de nombres de columnas a convertir a booleano
columnas_a_convertir = ['ADSL', 'Cablemódem', 'Dial Up', 'Fibra óptica', '4G', '3G', 'Telefonía Fija', 'Wireless', 'Satelital']

# Reemplazar "SI" por True y el resto de valores por False en las columnas seleccionadas, aqui se crean
# tipos de datos booleanos para trabajar mejor
for columna in columnas_a_convertir:
    df3I[columna] = df3I[columna].replace({'SI': True, np.nan: False})

df3I = df3I.reset_index(drop=True)

print(df3I.head(10))

# Define la ruta donde quieres guardar el archivo CSV
ruta_csv3 = 'Listado de localidades con conectividad a internet_cleaned.csv'  # Reemplaza con la ruta y el nombre que desees

# Guarda el DataFrame en un archivo CSV
df3I.to_csv(ruta_csv3, index=False)



# Se reviso si habian duplicados, datos anidados, valores nulos o NAN y no se evidencio nada.
```

```python
# Define la ruta completa del archivo CSV
Pintf_100H = 'Data_Acceso_Internet\Data_Acceso_Internet_Archivos_ Iniciales\Penetración de Internet fijo (accesos por cada 100 hogares).csv'  # Reemplaza 'ruta/a/tu/archivo.csv' con la ruta real

df4I = pd.read_csv(Pintf_100H)

# Verificar si hay valores nulos en el DataFrame
nulos_en_dataframe = df4I.isnull().any().any()

# Si hay valores nulos, puedes eliminar las columnas que contienen solo NaN:
df4I = df4I.dropna(axis=1, how='all')

# Y luego eliminar las filas que contienen NaN:
df4I = df4I.dropna()

# Verificar si hay filas duplicadas en el DataFrame
duplicados = df4I.duplicated().any()
print("¿Hay filas duplicadas en el DataFrame?:", duplicados)

# Convertir 'Año' a fecha:
df4I['Año'] = pd.to_datetime(df4I['Año'], format='%Y')


# Convertir valor float a entero
df4I['Accesos por cada 100 hogares'] = df4I['Accesos por cada 100 hogares'].str.replace(',', '.', regex=True).astype(float).astype(int)


print(df4I.head(10))

# Define la ruta donde quieres guardar el archivo CSV
ruta_csv4 = 'Penetración de Internet fijo (accesos por cada 100 hogares)_cleaned.csv'  # Reemplaza con la ruta y el nombre que desees

# Guarda el DataFrame en un archivo CSV
df4I.to_csv(ruta_csv4, index=False)

```

```python
# Define la ruta completa del archivo CSV
Phn_If = 'Data_Acceso_Internet\Data_Acceso_Internet_Archivos_ Iniciales\Penetración por hogares nacional de Internet fijo.csv'  # Reemplaza 'ruta/a/tu/archivo.csv' con la ruta real

df5I = pd.read_csv(Phn_If)

# Elimina la columna 'Periodo'
df5I = df5I.drop('Periodo', axis=1)

# Cambia el tipo de dato de la columna 'Año' a datetime
df5I['Año'] = pd.to_datetime(df5I['Año'], format='%Y')

# Convierte las columnas 'Accesos por cada 100 hogares' y 'Accesos por cada 100 hab' a enteros
df5I['Accesos por cada 100 hogares'] = df5I['Accesos por cada 100 hogares'].str.replace(',', '', regex=True).astype(float).round(0).astype(int)
df5I['Accesos por cada 100 hab'] = df5I['Accesos por cada 100 hab'].str.replace(',', '', regex=True).astype(float).round(0).astype(int)

duplicates = df5I[df5I.duplicated(keep=False)]
if duplicates.empty:
    print("No hay filas duplicadas en el DataFrame.")
else:
    print("Se encontraron filas duplicadas en el DataFrame:")
    print(duplicates)

print(df5I.head(10))

# Define la ruta donde quieres guardar el archivo CSV
ruta_csv5 = 'Penetración por hogares nacional de Internet fijo_cleaned.csv'  # Reemplaza con la ruta y el nombre que desees

# Guarda el DataFrame en un archivo CSV
df5I.to_csv(ruta_csv5, index=False)
```

ETL Telefonia Movil

```python
# Define la ruta completa del archivo CSV
Ingresos_trimestrales_pstm = 'Data_Telefonia_Movil\Data_Telefonia_Movil_Archivos_Iniciales\Ingresos trimestrales por la prestacion del servicio de telefonia movil.csv'  # Reemplaza 'ruta/a/tu/archivo.csv' con la ruta real

# Convertimos nuestro archivo en un df
df1m = pd.read_csv(Ingresos_trimestrales_pstm)

# Se Verificaron si habian valores nulos o NaN en el DataFrame, no se encontraron

# Se Verificaron si habian duplicados, no se encontraron

# Se Verificaron si habian valores en blanco, no se encontraron

# Eliminar la columna 'Periodo'
df1m = df1m.drop('Periodo', axis=1)

# Convertir la columna 'Año' a tipo de dato datetime
df1m['Año'] = pd.to_datetime(df1m['Año'], format='%Y')

# Quitar el símbolo '$', los puntos y el espacio en la columna 'Ingresos (miles de $)' y convertir a enteros
df1m['Ingresos (miles de $)'] = df1m['Ingresos (miles de $)'].str.replace('[$. ]', '', regex=True).astype(int)

# Convertir la columna 'Trimestre' a enteros
df1m['Trimestre'] = df1m['Trimestre'].astype(int)

print(df1m)

# Guardar el DataFrame modificado en un archivo CSV
df1m.to_csv('Ingresos_trimestrales_telefonia_movil_cleaned.csv', index=False)
```

```python
# Define la ruta completa del archivo CSV
Penetracion_ntm_acc_100_h = 'Data_Telefonia_Movil\Data_Telefonia_Movil_Archivos_Iniciales\Penetración nacional de la telefonía móvil (accesos por cada 100 habitantes).csv'  # Reemplaza 'ruta/a/tu/archivo.csv' con la ruta real

# Convertimos nuestro archivo en un df
df2m = pd.read_csv(Penetracion_ntm_acc_100_h)

# df2m.isnull().sum() no se identificaron datos nulos

# df2m.duplicated().sum() no se identificaron datos duplicados

df2m.drop('Periodo', axis=1, inplace=True) # Se elimina la columna Periodo

# Redondear y convertir la columna 'Accesos por cada 100 hab' a enteros
df2m['Accesos por cada 100 hab'] = df2m['Accesos por cada 100 hab'].round().astype(int)

# Convertir la columna 'Trimestre' a enteros
df2m['Trimestre'] = df2m['Trimestre'].astype(int)

# Convertir la columna 'Año' a tipo de dato datetime
df2m['Año'] = pd.to_datetime(df2m['Año'], format='%Y')

print(df2m)

# Guardar el DataFrame modificado en un archivo CSV
df2m.to_csv('Penetracion_Nacional_Telefonia_Movil_(accesos por cada 100 habitantes)_cleaned.csv', index=False)

```

```python
# Define la ruta completa del archivo CSV
Total_nac_acc_tel_mov = 'Data_Telefonia_Movil\Data_Telefonia_Movil_Archivos_Iniciales\Total nacional de accesos a telefonía móvil (trimestral).csv'  # Reemplaza 'ruta/a/tu/archivo.csv' con la ruta real

# Convertimos nuestro archivo en un df
df3m = pd.read_csv(Total_nac_acc_tel_mov)

df3m = df3m.drop(columns=['Periodo']) # Eliminamos Periodo

df3m['Año'] = pd.to_datetime(df3m['Año'], format='%Y') # Cambiamos la año a tipo de dato datetime

df3m['Trimestre'] = df3m['Trimestre'].astype(int) # Cambiamos la columna 'Trimestre' a tipo de datos entero

df3m[df3m.isna().any(axis=1)] # Se verifica si hay filas con valores nulos.

df3m = df3m.fillna(0)  # Se reemplazan  los valores nulos con ceros.

df3m[df3m.duplicated()] # Se verifican si hay filas duplicadas en el DataFrame.

# Cambiar la columna 'Trimestre' a tipo de dato entero
df3m['Trimestre'] = df3m['Trimestre'].astype(int)

# Cambiar todas las columnas de accesos a tipo de dato entero
columnas_accesos = ['Total de accesos pospago', 'Total de accesos prepago', 'Total de accesos operativos']
df3m[columnas_accesos] = df3m[columnas_accesos].apply(lambda x: x.str.replace('.', '').astype(int))

print(df3m)

# Guardar el DataFrame modificado en un archivo CSV
df3m.to_csv("Total_nacional de accesos a telefonía móvil (trimestral)_cleaned.csv", index=False)

```
