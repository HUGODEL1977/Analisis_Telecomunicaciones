Matriz de correlación:
Matriz de correlación para ver cómo las variables booleanas se relacionan entre sí.

Aca se revisan los tipos de tecnologia para conectividad a internet por provincia y localidad y se identifican por medio de datos booleanos que la tecnologia que mas se utiliza en la totalidad de provincias y localidades es 4G Y WIRELESS.

```python
# Verifica si hay valores nulos en las columnas booleanas
null_counts = df3.iloc[:, 2:].isnull().sum()

# Encuentra columnas con valores no válidos
columns_with_invalid_data = null_counts[null_counts > 0].index.tolist()

# Limpia los valores no válidos (por ejemplo, convierte "True" en True)
for column in columns_with_invalid_data:
    df3[column] = df3[column].replace({"True": True, "False": False})

# Genera los gráficos
fig, axes = plt.subplots(4, 3, figsize=(15, 10))
axes = axes.ravel()

for i, col in enumerate(df3.columns[2:]):
    sns.countplot(x=col, data=df3, ax=axes[i])
    axes[i].set_title(f'Distribución de {col}')
    axes[i].set_xticklabels(['False', 'True'])

plt.tight_layout()
plt.show()
```

Se pueden observar las tecnologias mas usadas.

```python
# Gráfico 3: Proporción de tecnologías en todas las provincias
plt.subplot(3, 2, 3)
tech_columns = df3.columns[2:]
tech_counts = df3[tech_columns].sum()
tech_counts.plot(kind='bar', color='skyblue')
plt.title('Proporción de Tecnologías en Todas las Provincias')
plt.xlabel('Tecnología')
plt.ylabel('Conteo')
```

Aqui encontramos la correlacion entre las tecnologias, si en el mapa de calor hay un color rojo (correlación positiva) o azul (correlación negativa), eso indica una relación más fuerte o mas debil entre las tecnologías. 

Se Puede identificar qué tecnologías están altamente relacionadas y podrían influirse mutuamente.

```python
# Gráfico 4: Mapa de calor de correlación
plt.figure(figsize=(12, 10))  # Ajusta el tamaño de la figura
plt.subplot(3, 2, 4)
corr_matrix = df3[tech_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=False)  # Elimina la barra de colores
plt.title('Matriz de Correlación de Tecnologías')
```

Aca tenemos el analisiS EDA y el KPI Propuesto, se aumento para el trimestre nuevo un 20 % ya que con el 2 % no se evidencia una variabilidad notable, podemos visualizar el resultado en el codigo:

Total de accesos por cada 100 hogares en el tercer trimestre de 2022: 1633
Total de accesos por cada 100 hogares en el cuarto trimestre de 2022 (con 20% de aumento): 1959.6

Y podemos observar el grafico con el KPI tambien propuesto.

```python
# Especifica la ruta del archivo CSV
archivo_csv4 = "Data_Acceso_Internet\Data_Acceso_Internet_Archivos_Limpios\Penetración de Internet fijo (accesos por cada 100 hogares)_cleaned.csv"

# Cargar el archivo CSV en un DataFrame
df4 = pd.read_csv(archivo_csv4)

# Configurar la visualización para mostrar todas las columnas en una fila
pd.set_option('display.width', 1000)  # Ajusta el ancho de visualización según tus necesidades

print(df4)

# KPI Propuesto medir el KPI propuesto a continuación, Aumentar en un 20% el acceso al servicio de internet para el 
# próximo trimestre, cada 100 hogares, por provincia. La fórmula es la siguiente: KPI 
# = ((Nuevoacceso - Accesoactual)/Accesoactual) * 100 

# Calcular el valor del "Nuevo acceso" para el cuarto trimestre de 2022, aumentando en un 20% el "Acceso actual" con base en
# (el tercer trimestre de 2022).

# Filtra los datos del tercer trimestre de 2022
tercer_trimestre = df4[(df4['Año'] == '2022-01-01') & (df4['Trimestre'] == 3)]

# Calcula el total de accesos por cada 100 hogares para el tercer trimestre
total_tercer_trimestre = tercer_trimestre['Accesos por cada 100 hogares'].sum()

# Calcula el total de accesos con el 20% de aumento para el cuarto trimestre
total_cuarto_trimestre = total_tercer_trimestre * 1.20

# Imprime los totales
print("Total de accesos por cada 100 hogares en el tercer trimestre de 2022:", total_tercer_trimestre)
print("Total de accesos por cada 100 hogares en el cuarto trimestre de 2022 (con 20% de aumento):", total_cuarto_trimestre)

# Filtra los datos del tercer trimestre de 2022
trimestre_actual = df4[(df4['Año'] == '2022-01-01') & (df4['Trimestre'] == 3)].copy()

# Calcula el nuevo acceso con el aumento del 2%
trimestre_actual['NuevoAcceso'] = trimestre_actual['Accesos por cada 100 hogares'] * 1.20

# Combina los DataFrames del tercer y cuarto trimestre de 2022
trimestre_cuarto = df4[(df4['Año'] == '2022-01-01') & (df4['Trimestre'] == 4)]
trimestres = pd.concat([trimestre_actual, trimestre_cuarto], ignore_index=True)

# Calcula el KPI
trimestres['KPI'] = ((trimestres['NuevoAcceso'] - trimestres['Accesos por cada 100 hogares']) / trimestres['Accesos por cada 100 hogares']) * 100

# Genera el gráfico
plt.figure(figsize=(12, 6))
plt.bar(trimestres['Provincia'], trimestres['Accesos por cada 100 hogares'], label='Tercer Trimestre 2022', alpha=0.7)
plt.bar(trimestres['Provincia'], trimestres['NuevoAcceso'], label='Cuarto Trimestre 2022 (20% de aumento)', alpha=0.7)
plt.xlabel('Provincia')
plt.ylabel('Accesos por cada 100 hogares')
plt.title('Acceso a Internet por Provincia en el Tercer y Cuarto Trimestre de 2022')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()

plt.show()

```

Para "Trimestre", el valor promedio es 2.5, lo que significa que los trimestres están distribuidos de manera bastante uniforme entre 1 y 4. "Tasa de Crecimiento Anual de los Accesos Operativos", Este KPI dará una idea de cómo han evolucionado los accesos operativos año tras año.

Este es el paso a paso:

Se  Agrupan los datos por año.

Se calcula la suma total del "Total de accesos operativos" para cada año.

Se calcula la tasa de crecimiento anual utilizando la fórmula:

TasadeCrecimientoAnual= ((AccesosOperativosdelActual−AccesosOperativosdelAoAnterior)/AccesosOperativosdelAoAnterior)*100
​

```python
# Agrupa los datos por año y calcula la suma de accesos operativos para cada año
df3m['Año'] = pd.to_datetime(df3m['Año']).dt.year  # Convierte la columna 'Año' a tipo datetime y obtiene el año
accesos_anuales = df3m.groupby('Año')['Total de accesos operativos'].sum()

# Calcula la tasa de crecimiento anual
tasa_crecimiento_anual = accesos_anuales.pct_change() * 100


tasa_crecimiento_anual = tasa_crecimiento_anual.drop(2013, errors='ignore')

# Imprime el resultado
print(tasa_crecimiento_anual)
```

```python
# Crea un gráfico de barras con colores personalizados
colores = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'teal', 'gray', 'cyan']

plt.figure(figsize=(10, 6))
plt.bar(tasa_crecimiento_anual.index, tasa_crecimiento_anual, color=colores)
plt.xlabel('Año')
plt.ylabel('Tasa de Crecimiento Anual (%)')
plt.title('Tasa de Crecimiento Anual de Accesos Operativos (2014-2022)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Muestra el gráfico
plt.show()
```

Se revisa la evolucion en el acceso de telefonia movil pospago, prepago y la telefonia movil en general, podemos identificar un patron en el año 2019 - 2020 donde al crecer la telefonia movil pospago, la prepago tiene un efecto contrario y viene en bajada desde el año 2017 hasta el año 2020, es importante aclarar que en el año 2020 pasa lo contrario, se desploman los accesos a la telefonia pospago y se incrementan los accesos a la telefonia prepago, esto tiene su efecto en la pandemia covid 19 y en las perdidas de empleo que limitan el poder adquisitivo de las personas y deben buscar una reduccion en sus gastos, por ende se pasan a una telefonia mas economica como lo es la telefonia prepago, los accesos operativos tienen su mayor variabilidad en los años 2019 - 2020 - 2021 por lo anteriormente mencionado.
